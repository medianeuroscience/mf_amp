```python
import pandas as pd 
pd.options.display.max_rows = 400
```

```python
moral_words = pd.read_pickle('../data/moral_words.pkl')
moral_words['category'] = moral_words.foundation
moral_words = moral_words['category']

neutral_words = pd.read_pickle('../data/neutral_words.pkl')
neutral_words['category'] = 'neutral'
neutral_words = neutral_words['category']

non_words = pd.read_pickle('../data/nonwords.pkl')
non_words['category'] = 'nonword'
non_words = non_words['category']
```

```python
all_words = pd.concat([moral_words,neutral_words,non_words])
```

```python
all_words.value_counts()
```

## Words Per Task

Lexical Decision Task:
- 10 per foundation (5 virtue and 5 vice per foundation) = 50 moral words
- 50 Neutral words
- 100 Non-words

Affect Misattribution:
- 14 per foundation = 70 moral words
- 70 non-words

Total:
- 120 moral words
- 50 neutral words
- 170 nonwords

```python
stimuli = pd.DataFrame(all_words)
```

```python
stimuli['correct_amp'] = 'nan'
stimuli['correct_ldt'] = 'nan'
```

```python
for i,row in stimuli.iterrows():
    cat = row.category
    if cat.endswith('.vice'):
        stimuli.at[i, 'correct_amp'] = 'r'
        stimuli.at[i, 'correct_ldt'] = 'l'
    elif cat.endswith('.virtue'):
        stimuli.at[i, 'correct_amp'] = 'l' 
        stimuli.at[i, 'correct_ldt'] = 'l'
    elif cat == 'nonword':
        stimuli.at[i, 'correct_ldt'] = 'r'
    elif cat == 'neutral':
        stimuli.at[i, 'correct_ldt'] = 'l'
```

```python
stimuli['word'] = stimuli.index
stimuli = stimuli.reset_index()
del stimuli['index']
```
